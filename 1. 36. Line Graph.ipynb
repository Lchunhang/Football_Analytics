```python
#Import Packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from adjustText import adjust_text
import matplotlib.font_manager as fm
from highlight_text import  ax_text, fig_text

df = pd.read_csv('LineChart.csv')
df.head()
```

```python
#make the plot
fig, ax = plt.subplots(figsize=(30,20))
fig.set_facecolor('black')
mpl.rcParams['figure.dpi'] = 400

#set up our base layer
mpl.rcParams['xtick.color'] = 'yellow'
mpl.rcParams['ytick.color'] = 'yellow'

#background color
ax.patch.set_facecolor('black')

#setting up the grid
ax.grid(ls='-.', lw=3, color='white', axis='y', zorder=0, alpha=0.2)
ax.grid(ls='-.', lw=3, color='white', axis='x', zorder=0, alpha=0.2)

#Inverse the axis
#plt.gca().invert_yaxis()

#Setting up the spine
spines = ['top','right','bottom','left']
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)
        
#Set spine width & color
ax.spines['left'].set_linewidth(7)
ax.spines['bottom'].set_linewidth(7)
ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')

#set the ticks

ax.set_xticks([1,2,3])

ax.set_yticks([30, 40, 50, 60])

ax.set_ylabel('Points',fontname='Courier New',color='white',
              fontsize=60, fontweight = 'bold', labelpad = 30)

ax.set_xlabel('Season',fontname='Courier New',color='black',fontsize=40, 
              fontweight = 'bold', labelpad = 40)

#Add padding between axes and labels
ax.xaxis.set_tick_params(pad=10, width=0, length=0, direction='in')
ax.yaxis.set_tick_params(pad=10, width=0, length=5)

# change the fontsize
ax.tick_params(axis='x', labelsize=0)

# change the fontsize
ax.tick_params(axis='y', labelsize=40, labelcolor='white')

#Plot Line
ax.plot(df['Season'], df['Points'],lw=15, color='w',zorder=-2)

#Plot Markers
ax.plot(df['Season'], df['Points'], ms = 92, mec = 'w', 
        mew=8, lw=0, marker='o', color='#0033CC')

#set title
highlight_textprops =\
    [{"color": 'yellow',"fontweight": 'bold'}]

Title = fig_text(s = "<Hoffenheim's> Points under Julian Nagelsmann",
        x = 0.5, y = 0.99, color='white',highlight_textprops = highlight_textprops,
        fontfamily = "Courier New", fontweight='bold', ha='center',
        va= 'center',fontsize=60)

fig_text(s = 'Bundesliga (2015/16 - 2018/19)', x=0.5, y=0.95, color= 'w',fontsize=60,
         fontfamily = "Courier New", fontweight='bold', ha='center')

#Space
ax.text(2, 34, 'space', va='center', ha='left', fontsize=25, color='black',
        fontweight = "bold",fontfamily = "Courier New")

#shade
plt.axvspan(1,2, ymax=0.98, color='lightgrey',zorder=-3,alpha=0.5)
plt.axvspan(2,3, ymax=0.98, color='#0060a9',zorder=-3,alpha=0.5)
plt.axvspan(3,4, ymax=0.98, color='lightgrey',zorder=-3,alpha=0.5)

#Draw line
ax.axvline(2, 0, 0.98, color='yellow', ls='--',lw=8, zorder=-2, alpha= 0.9)
ax.axvline(3, 0, 0.98, color='yellow', ls='--',lw=8, zorder=-2, alpha= 0.9)

#Annotations
plt.text(1, 37, '37', fontsize=35, color='w', fontweight = "bold",fontfamily = "Courier New", va='center',ha='center')
plt.text(2, 62, '62', fontsize=35, color='w', fontweight = "bold",fontfamily = "Courier New", va='center',ha='center')
plt.text(3, 55, '55', fontsize=35, color='w', fontweight = "bold",fontfamily = "Courier New", va='center',ha='center')
plt.text(4, 51, '51', fontsize=35, color='w', fontweight = "bold",fontfamily = "Courier New", va='center',ha='center')

"""
#Annotations
plt.text(2.85, 43, 'Champions League\nQualification or Play', fontsize=50, color='w', alpha=0.8, rotation='vertical', zorder = -1,
         fontweight = "bold",fontfamily = "Courier New", va='center',ha='center')
"""
#Annotations
plt.text(2.5, 43, 'Champions League\nQualification\nor Playoffs', fontsize=48, color='w', alpha=0.8, zorder = -1,
         fontweight = "bold",fontfamily = "Courier New", va='center',ha='center')
plt.show()
```
